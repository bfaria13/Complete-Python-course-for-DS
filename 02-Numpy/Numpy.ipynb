```python
# Realizando operações de multiplicação
start_time_numpy = time.time()

%time resultado_numpy = array * array

end_time_numpy = time.time()
print("Tempo de execução usando arrays:", end_time_numpy - start_time_numpy)
```

```python
# Realizando operações de multiplicação
start_time_lista = time.time()

%time resultado_lista = [x * x for x in lista]

end_time_lista = time.time()
print("Tempo de execução usando listas Python:", end_time_lista - start_time_lista)
```

```python
%%time

# Realizando operações de multiplicação
start_time_numpy = time.time()

resultado_numpy = array * array

end_time_numpy = time.time()
print("Tempo de execução usando arrays:", end_time_numpy - start_time_numpy)
```

```python
%%time

# Realizando operações de multiplicação
start_time_lista = time.time()

resultado_lista = [x * x for x in lista]

end_time_lista = time.time()
print("Tempo de execução usando listas Python:", end_time_lista - start_time_lista)
```

<a id="s6a"><h3> 6.1 Estatística descritiva </h3></a>

```python
# Cálculo de média, mediana, desvio padrão, minimo e máximo para listas e arrays

lista = [1,2,3,4,5]
array = np.array(lista)
print('lista:', lista, ' array:', array)
print()

mean = np.mean(lista)
median = np.median(lista)
std = np.std(lista)
vmin = np.min(lista)
vmax = np.max(lista)

print(f'média: {mean}, mediana: {median}, desvio padrão: {std}, min: {vmin} e max: {vmax} para a lista')
print()

mean = np.mean(array)
median = np.median(array)
std = np.std(array)
vmin = np.min(array)
vmax = np.max(array)

print(f'média: {mean}, mediana: {median}, desvio padrão: {std}, min: {vmin} e max: {vmax} para o array')
print()
```

```python
# Cálculo de média, mediana, desvio padrão, minimo e máximo para matrizes

matriz = np.asmatrix([[1,2,3],[4,5,6],[7,8,9]])
print('matriz:\n', matriz, ' dimensão:', matriz.shape)
print()

mean = np.mean(matriz)
median = np.median(matriz)
std = np.std(matriz)
vmin = np.min(matriz)
vmax = np.max(matriz)

print(f'média: {mean}, mediana: {median}, desvio padrão: {std}, min: {vmin} e max: {vmax} para a matrix')
print()
```

#### Trabalhe o código para aplicá-lo ao tensor. Como podemos realizar cada uma das operações sobre um tensor 3 dimensional?

```python
# Para calcular as medidas para matrizes é importante específicar sobre qual eixo deve-se aplicar o cálculo!!
# axis=0 aplica o calculo na coluna; axis=1 aplica o cálculo na linha

matriz = np.asmatrix([[1,2,3],[4,5,6],[7,8,9]])
print('matriz:\n', matriz, ' dimensão:', matriz.shape)
print()

mean = np.mean(matriz, axis=0)
median = np.median(matriz, axis=0)
std = np.std(matriz, axis=0)
vmin = np.min(matriz, axis=0)
vmax = np.max(matriz, axis=0)

print(f'média: {mean}, mediana: {median}, desvio padrão: {std}, min: {vmin} e max: {vmax} para cada coluna da matrix')
print()

mean = np.mean(matriz, axis=1)
median = np.median(matriz, axis=1)
std = np.std(matriz, axis=1)
vmin = np.min(matriz, axis=1)
vmax = np.max(matriz, axis=1)

print(f'média:\n{mean}\nmediana:\n{median}\ndesvio padrão:\n{std}\nmin:\n{vmin}\ne max:\n{vmax}\npara cada linha da matrix')
print()
```

<a id="s6b"><h3> 6.2 Soma, produto e acumulados </h3></a>

```python
# Cálculo da soma e produtos, e também os acumulados para listas e vetores

lista = [1,2,3,4,5]
array = np.array(lista)
print('lista:', lista, ' array:', array)
print()

soma = np.sum(lista)
somaacum = np.cumsum(lista)
prod = np.prod(lista)
prodacum = np.cumprod(lista)


print(f'soma: {soma}, soma acumulada: {somaacum}, produto: {prod}, produto acumulado: {prodacum} para a lista')
print()

soma = np.sum(array)
somaacum = np.cumsum(array)
prod = np.prod(array)
prodacum = np.cumprod(array)


print(f'soma: {soma}, soma acumulada: {somaacum}, produto: {prod}, produto acumulado: {prodacum} para o array')
print()
```

```python
# Cálculo da soma e produtos, e também os acumulados para matrizes

matriz = np.asmatrix([[1,2,3],[4,5,6],[7,8,9]])
print('matriz:\n', matriz, ' dimensão:', matriz.shape)
print()

soma = np.sum(matriz)
somaacum = np.cumsum(matriz)
prod = np.prod(matriz)
prodacum = np.cumprod(matriz)

print(f'soma: {soma}, soma acumulada: {somaacum}, produto: {prod},\nproduto acumulado: {prodacum} para a matriz')
print()
```

```python
# Para calcular as medidas para matrizes é importante específicar sobre qual eixo deve-se aplicar o cálculo!!
# axis=0 aplica o calculo na coluna; axis=1 aplica o cálculo na linha

matriz = np.asmatrix([[1,2,3],[4,5,6],[7,8,9]])
print('matriz:\n', matriz, ' dimensão:', matriz.shape)
print()

soma = np.sum(matriz, axis=0)
somaacum = np.cumsum(matriz, axis=0)
prod = np.prod(matriz, axis=0)
prodacum = np.cumprod(matriz, axis=0)

print(f'soma:\n{soma}, soma acumulada:\n{somaacum}, produto:\n{prod},\nproduto acumulado:\n{prodacum}\npara cada coluna da matriz')
print()

soma = np.sum(matriz, axis=1)
somaacum = np.cumsum(matriz, axis=1)
prod = np.prod(matriz, axis=1)
prodacum = np.cumprod(matriz, axis=1)

print(f'soma:\n{soma}\nsoma acumulada:\n{somaacum}\nproduto:\n{prod}\nproduto acumulado:\n{prodacum}\npara cada linha da matrix')
print()
```

<a id="s6c"><h3> 6.3 Reshaping, sorting e embaralhamento </h3></a>

```python
# Reshaping de arrays

array = np.array([1,2,3,4,5,6])

print('array:\n', array, ' dimensão:', array.shape)
print()

# matriz com 1 linha e 6 colunas
m1_6 = array.reshape(1,-1) # 1: uma linha, -1: todos os elementos do array
print('matriz:\n', m1_6, ' dimensão:', m1_6.shape)
print()

# matriz com 3 linha e 2 colunas
m3_2 = array.reshape(3, 2)
print('matriz:\n', m3_2, ' dimensão:', m3_2.shape)
print()

# Transformando em uma única dimensão novamente: função flatten()
array_flattened = m3_2.flatten()
print("Array:\n", array_flattened, ' dimensão:', array_flattened.shape)
```

```python
# Reshaping de matrizes

matriz = np.asmatrix([[1,2,3],[4,5,6],[7,8,9]])

print('matriz:\n', matriz, ' dimensão:', matriz.shape)
print()

# matriz com 1 linha e 9 colunas
m1_6 = matriz.reshape(1,-1) # 1: uma linha, -1: todos os elementos do array
print('matriz:\n', m1_6, ' dimensão:', m1_6.shape)
print()

# matriz com 3 linha e 3 colunas
m3_3 = matriz.reshape(3, 3)
print('matriz:\n', m3_3, ' dimensão:', m3_3.shape)
print()

# Transformando em uma única linha: função flatten()
matriz_flattened = m3_3.flatten()
print("matriz:\n", matriz_flattened, ' dimensão:', matriz_flattened.shape)

# Transformando em um array unidimensional: funções squeeze() e asarray()
array_flattened = np.squeeze(np.asarray(matriz_flattened))
print("Array:\n", array_flattened, ' dimensão:', array_flattened.shape)
```

<a id="s6e"><h3> 6.5 Geração de sequências numéricas </h3></a>

```python
# Sequência de números usando np.arange() 

sequencia = np.arange(start=1, stop=8, step=0.5)  # [start, stop)
print("Sequência numérica usando np.arange():", sequencia)
print()

sequencia = np.arange(start=0, stop=100, step=10)  # [start, stop)
print("Sequência numérica usando np.arange():", sequencia)
print()

sequencia = np.arange(start=0, stop=100, step=10, dtype=float)  # [start, stop)
print("Sequência numérica usando np.arange():", sequencia)
print()
```

```python
# Sequência de números usando np.linspace() 

sequencia = np.linspace(start=1, stop=8, num=10, endpoint=True)  # [start, stop]
print("Sequência numérica usando np.linspace():\n", sequencia)
print()

sequencia = np.linspace(start=1, stop=8, num=10, endpoint=False)  # [start, stop)
print("Sequência numérica usando np.linspace():\n", sequencia)
print()

sequencia = np.linspace(start=1, stop=8, num=10, endpoint=True, dtype=int)  # [start, stop]
print("Sequência numérica usando np.linspace():\n", sequencia)
print()
```

<a id="s6i"><h3> 6.9 Tratamento de valores nulos </h3></a>

```python
# Criando um array NumPy com valores NaN
array_nan = np.array([1, np.nan, 3, np.nan, 5])
print('Array com nulo', array_nan)
print()

# Verificando quais elementos são NaN
valores_nan = np.isnan(array_nan)
print("Quais elementos são NaN:", valores_nan)
print()

# Determinando o número de NaNs
numero_nans = np.count_nonzero(valores_nan)
print("Número de NaNs:", numero_nans)
print()

# Calculando a média do array ignorando NaNs
media_sem_nans = np.nanmean(array_nan)
print("Média do array ignorando NaNs:", media_sem_nans)

# # Calculando o máximo e o mínimo do array ignorando NaNs
maximo_sem_nans = np.nanmax(array_nan)
minimo_sem_nans = np.nanmin(array_nan)
print("Máximo do array ignorando NaNs:", maximo_sem_nans)
print("Mínimo do array ignorando NaNs:", minimo_sem_nans)
```

```python
# Substituição de NaNs pela média ou qualquer outra medida

# Calculando a média dos valores não nulos
media = np.nanmean(array_nan)

# Substituindo os valores nulos pela média
array_sem_nulos = np.where(np.isnan(array_nan), media, array_nan)

print('array original: ', array_nan)
print('array tratado: ', array_sem_nulos)
```

<a id="s6j"><h3> 6.10 Tratamento de valores infinito </h3></a>

```python
# Substituição dos infinitos pelo valor 0 ou qualquer outra medida

array_inf = np.array([1., np.inf, -np.inf, 2.])
array_sem_inf = array_inf.copy()

array_sem_inf[(np.isinf(array_sem_inf) | (np.isneginf(array_sem_inf)))] = 0

print('array original: ', array_inf)
print('array tratado: ', array_sem_inf)
```
